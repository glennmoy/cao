```python
import numpy as np
import pandas as pd
import pandasql as ps
import csv
import scipy.stats
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.text

# I'm going to use seaborn palette
current_palette=sns.color_palette()

# Import data
# [code year points college audition]
cao_data=pd.read_csv('course_points_col.csv')
# [year points num]
cand_data=pd.read_csv('awarded_points_col.csv')
```

```python
# Function to perform linear regression on x_data, y_data
# Print graph? y/n
def do_regression(x_data,y_data,printyn):

    slope, intercept, r_value, p_value, std_err = \
    scipy.stats.linregress(x_data,y_data)
    
    predicted_y=slope*x_data+intercept
    pred_error=y_data-predicted_y
    dof=len(x_data)-2
    res_std_err=np.sqrt(np.sum(pred_error**2)/dof)

    # Print figure if printyn==True
    if printyn:
        print("\nslope\t\t=",slope)
        print("intercept\t=",intercept)
        print("r_value\t\t=",r_value)
        print("p_value\t\t=",p_value)
        print("std_err\t\t=",std_err)

        f=plt.figure()
        ax=f.add_subplot(111)
        textstr = '$y=%.2f+%.2fx$\n$r=%.2f$'%(intercept, slope, r_value)
        ax.text(0.05, 0.95, textstr, transform=ax.transAxes, fontsize=14,\
                verticalalignment='top',backgroundcolor='1.0')

        plt.scatter(x_data,y_data)
        plt.plot(x_data,predicted_y,'r-')
        plt.savefig('regression.pdf',bbox_inches='tight')
        plt.show()

    return slope,intercept,r_value,p_value,res_std_err
```

```python
# Function to perform auto-regression on a set of data
# with an interval of 1 year
def auto_regress(df,year1,year2,printyn):

    old_points=[]
    new_points=[]

    for y in range(year1,year2):
        old_points.extend(df[str(y)].values)
        new_points.extend(df[str(y+1)].values)

    old_points=np.array(old_points)
    new_points=np.array(new_points)

    slope, intercept,r_value,p_value,std_err =\
    do_regression(old_points,new_points,printyn)

    return slope,intercept
```

```python
# Normalise the points post 2012 to account for Maths bonus points
def normalise_points(factor,constant):
    
    for index,row in cao_data_norm.iterrows():
        if int(row['year'])>=2012 and not np.isnan(row['points']):
            p=cao_data_norm.iloc[index].points
            
            if p<=625:
                cao_data_norm.set_value(index,'points',(p-constant)/factor)
            else:
                cao_data_norm.set_value(index,'points',p-25)
    return cao_data_norm
```

```python
### PART 0: Normalising the data

#Query the points from 2001-2016
q0=""" SELECT code,points,year FROM cao_data \
        WHERE year>=2001 AND audition==0;"""
sample_data=ps.sqldf(q0,locals())
sample_data.columns=['code','points','year']
sample_data=sample_data.pivot(index='code',columns='year',\
                              values='points').dropna()

sample_data.columns=['2001','2002','2003','2004','2005','2006',\
                     '2007','2008','2009','2010','2011','2012',\
                     '2013','2014','2015','2016']
sample_data.index.name='code'


# Large slope means higher courses are affected more
# slope = (0.915,1.006) -> Change in points = ((-0.06,8.5),(-3.6,50))

# Large intercept means lower courses are affected more
# intercept = (-20,42)

# 2001-2016 : (0.96493439363586697, 15.93537064658284)
print("2001-2016")
print(auto_regress(sample_data,2001,2016,False))

# 2006-2012 : (0.95796703858667986, 20.084892111872591)
print("2001-2016")
print(auto_regress(sample_data,2006,2012,False))

#2012-2016 : (0.95639672971610457, 24.047529002643444)
print("2001-2016")
print(auto_regress(sample_data,2012,2016,False))

#2011-2013 : (0.98127282019399475, 14.456593554233677)
print("2011-2013")
print(auto_regress(sample_data,2011,2013,False))
print()

print("All years")
for year in range(2001,2016):
    print(str(year)+"-"+str(year+1),\
          auto_regress(sample_data,year,year+1,False))
     
# 2001-2002 (0.95827418864167901, 12.022932919486834)
# 2002-2003 (0.96202556957745577, 12.925657817855495)
# 2003-2004 (1.0068006537532399, -1.0494257617568792)
# 2004-2005 (0.95165766484213554, 15.142911516293964)
# 2005-2006 (0.98111867595618441, 0.16445968100430264)
# 2006-2007 (0.92832050438773961, 30.025070521187331)
# 2007-2008 (0.92169424588286397, 28.744911338414738)
# 2008-2009 (0.92231548826730692, 35.030095695720206)
# 2009-2010 (0.96246678231475324, 18.565979053994909)
# 2010-2011 (0.95787199708945836, 23.662243119014477)
# 2011-2012 (1.0625641648985011, -20.002919633222575)
# 2012-2013 (0.91566559712467754, 42.925100367125822)
# 2013-2014 (0.98642639922254838, 7.1424937235379389)
# 2014-2015 (0.94997742917980765, 32.086148859777552)
# 2015-2016 (0.98485764431773104, 8.608068173409265)
```

```python
# There could be an effect from number of LC students each year
# Try normalise for that points = points*num_lc_stu/average_num

#Doesn't seem to have much effect for the higher courses

#number of candidates per year
q1=""" SELECT year,SUM(num) FROM cand_data GROUP BY year;"""
num_cands_per_year=ps.sqldf(q1,locals())
num_cands_per_year.columns=['year','num']

sample_data_norm=sample_data.copy()

average_num_cands=num_cands_per_year['num'].mean()

for col in sample_data_norm:
    sample_data_norm[col]=(sample_data_norm[col]*\
    num_cands_per_year.transpose()[int(col)-2001]['num'])/average_num_cands


# 2001-2016 : (0.958571583988578, 18.033843342609998)
print("2001-2016")
print(auto_regress(sample_data_norm,2001,2016,False))

# 2006-2012 : (0.95525569973233826, 22.620820309000692)
print("2001-2016")
print(auto_regress(sample_data_norm,2006,2012,False))

#2012-2016 : (0.97235398713363341, 23.350480956989372)
print("2001-2016")
print(auto_regress(sample_data_norm,2012,2016,False))

#2011-2013 : (0.96409036271346815, 15.239821400186827)
print("2011-2013")
print(auto_regress(sample_data_norm,2011,2013,False))
print()

print("All years")
for year in range(2001,2016):
    print(str(year)+"-"+str(year+1),\
          auto_regress(sample_data_norm,year,year+1,False))
```

```python
### Normalise Points

# Since Maths Bonus points are more likely to affect higher courses
# and given that normalising didn't qualitatively change anything
# I'll use the auto-regression parameters from 2011-2012 since 
# it's only a small effect anyway.

factor,constant = auto_regress(sample_data,2011,2012,False)
cao_data_norm=cao_data.copy()
normalise_points(factor,constant)
```

```python
### PART 1: How do number of courses compare to number of LC students?

#number of courses per year
q2=""" SELECT year,COUNT(code) FROM cao_data_norm GROUP BY year;"""
num_courses_per_year=ps.sqldf(q2,locals())
num_courses_per_year.columns=['year','num']

#Make a graph
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax2=ax1.twinx()
ax1.set_ylim([50000,57000])
ax1.set_xticklabels(['2001','2003','2005','2007','2009','2011','2013','2015'])
ax1.set_xlabel('Year')
ax1.set_ylabel('Number of LC Students')

ax1=num_cands_per_year.num.plot(ax=ax1,marker='s',\
                                color=current_palette[1],\
                                label='Number of LC Candidates')

ax2=num_courses_per_year.num.plot(ax=ax2,marker='o',\
                                  color=current_palette[0],\
                                  label='Number of L8 Courses',\
                                  secondary_y=True)
ax2.set_ylim([300,1000])
ax2.set_ylabel('Number of Courses')
ax=plt.gca()
lines = ax1.get_lines() + ax2.get_lines() 
ax.legend(lines, [l.get_label() for l in lines],loc='upper center')
plt.savefig('num_cands_vs_num_courses.pdf',bbox_inches='tight')

# Also do a regression on analysis here
slope,intercept,r_value,p_value,std_err=\
do_regression(num_cands_per_year.num,num_courses_per_year.num,True)


```

```python
### PART 2: How do the course points compare to number of LC students?

#average points over all courses per year
q4=""" SELECT year,AVG(points) FROM cao_data_norm GROUP BY year;"""
av_points_per_year=ps.sqldf(q4,locals())
av_points_per_year.columns=['year','points']

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax2=ax1.twinx()
ax1.set_ylim([50000,59000])
ax1.set_xticklabels(['2001','2003','2005','2007','2009',\
                     '2011','2013','2015','2016'])
ax1.set_xlabel('Year')
ax1.set_ylabel('Number of LC Students')

ax1=num_cands_per_year.num.plot(ax=ax1,marker='s',color=current_palette[1],label='Number of LC Candidates')
#ax2=av_points_per_year.points.plot(ax=ax2,marker='^',color=current_palette[2],label='Average Course Points',secondary_y=True)
ax2=av_points_per_year.points.plot(ax=ax2,marker='>',color=current_palette[5],label='Average Course Points (Normalised)',secondary_y=True)
ax2.set_ylabel('Average Cut-off Points (Normalised)')
ax2.set_ylim([350,395])
ax=plt.gca()
lines = ax1.get_lines() + ax2.get_lines()
ax.legend(lines, [l.get_label() for l in lines],loc='upper center')
plt.savefig('num_cands_vs_av_points_norm.pdf',bbox_inches='tight')

slope, intercept,r_value,p_value,std_err=do_regression(num_cands_per_year.num,av_points_per_year.points,True)
```
