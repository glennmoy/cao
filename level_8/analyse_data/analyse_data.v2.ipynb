```python
# Function to perform linear regression on x_data, y_data
# Print graph? y/n
def do_regression(x_data,y_data,printyn,identity):

    slope, intercept, r_value, p_value, std_err = \
    scipy.stats.linregress(x_data,y_data)
    
    predicted_y=slope*x_data+intercept
    pred_error=y_data-predicted_y
    dof=len(x_data)-2
    res_std_err=np.sqrt(np.sum(pred_error**2)/dof)

    # Print figure if printyn==True
    if printyn:
        print("\nslope\t\t=",slope)
        print("intercept\t=",intercept)
        print("r_value\t\t=",r_value)
        print("p_value\t\t=",p_value)
        print("std_err\t\t=",std_err)

        f=plt.figure()
        ax=f.add_subplot(111)
        textstr = '$y=%.2f+%.2fx$\n$r=%.2f$'%(intercept, slope, r_value)
        ax.text(0.05, 0.95, textstr, transform=ax.transAxes, fontsize=14,\
                verticalalignment='top',backgroundcolor='1.0')

        plt.scatter(x_data,y_data)
        plt.plot(x_data,predicted_y,'r-')
        if(identity): plt.plot(x_data,x_data,'k-')
        plt.savefig('images/regression.pdf',bbox_inches='tight')
        plt.show()

    return slope,intercept,r_value,p_value,res_std_err
```

```python
# Function to perform auto-regression on a set of data
# with an interval of 1 year
def auto_regress(df,year1,year2,printyn,identity):

    old_points=[]
    new_points=[]

    for y in range(year1,year2):
        old_points.extend(df[str(y)].values)
        new_points.extend(df[str(y+1)].values)

    old_points=np.array(old_points)
    new_points=np.array(new_points)

    slope, intercept,r_value,p_value,std_err =\
    do_regression(old_points,new_points,printyn,identity)

    return slope,intercept
```

```python
### FINDINGS PART 1

# Despite an almost yearly increase in the number of courses available, 
# course points are generally increasing due to the ever-increasing 
# number of applicants.

# This is further reflected in the course points vs candidate points 
# which were uncorrelated before 2007.
# But once demand for places increased after the crash, demand 
# exceeded supply and the average course points are now highly 
# linked with student performance, which is steadily increasing.

# The number of places in each course is not taken into account and 
# I would expect this to have an effect. Most likely the many courses 
# are small which keeps demand high.

# The incrase in candidate points is increasing year-on-year and 
# cannot simply be attributed to Maths Bonus points. It begs the 
# question if students are getting more intelligent, better at 
# gaming the system, or if the exams are becoming easier.

# What this speaks of is a highly competitive application procedure 
# with Universities struggling to keep up with demand despite 
# providing more and more courses each year.
```

```python
# Function to plot college rankings

def plot_college_rankings(rankings_df,name):
    # Careful of where min/max applies to 2+ colleges!!

    #Find min and max course points for colleges with min/max mean/std
    max_mean_college=rankings_df.head(1)
    min_mean_college=rankings_df.tail(1)

    #histogram colors will change depending on range    
    mask1=rankings_df['mean'] < 500
    mask2=rankings_df['mean'] < 400
    mask3=rankings_df['mean'] < 300
    mask4=rankings_df['mean'] < 200
   
    colors = np.array([current_palette[0]]*len(rankings_df))
    colors[np.where(mask1)] = current_palette[1]
    colors[np.where(mask2)] = current_palette[2]
    colors[np.where(mask3)] = current_palette[3]
    colors[np.where(mask4)] = current_palette[4]
   
    ax1 = rankings_df.dropna().plot(x='college',y='mean',\
                                    kind='barh',figsize=(9,9.5),\
                                    legend=False,fontsize=12,grid=True,color=colors)
    
    ax1.set_title("Average Course Points per University",\
                  fontsize=16) # x-label
    ax1.set_xlabel("Mean Points",fontsize=12) # x-label
    ax1.set_ylabel("",fontsize=12)       # y-label
    ax1.tick_params(axis='y', pad=10)    # add padding bw y-tics and y-axis
    plt.gca().invert_yaxis()            # sort by descending order
  
    # Put the points on the right hand side 
    for idx,line in rankings_df.iterrows():
   
        for p in ax1.patches:
            mean=p.get_width()
            ax1.annotate(str(int(mean)),(600,p.get_y()+1.2*p.get_height()),\
                         fontsize=12)
    
    fig=plt.gcf()
    fig.subplots_adjust(left=0.5,right=0.85,top=0.95,bottom=0.08) # give more space to college names
    plt.savefig("images/"+name+".pdf")
    plt.show()
    
    #print(ggplot(av_points_per_college,aes(x='college',weight='points'))\
    #      +geom_bar(stat='identity')+coord_flip())
```

```python
# Function to get sorted college rankings

def get_college_rankings(audition,year):
    college_name=[];mean=[];std=[]

    if audition:
        # If specified, rank for a specific year
        # else rank over all years
        if year>2000:
            name=str(year)+"aud"
            for college in cao_data['college']\
            [(cao_data['year']==year)].dropna().unique():
                m=cao_data.dropna().points[(cao_data.college==college)\
                                                & (cao_data.year==year)].mean()
                s=0
                college_name.append(college); 
                mean.append(m); std.append(s)

        else:
            name="all_years_aud"
            for college in cao_data['college'].dropna().unique():
                m=cao_data.dropna().points[(cao_data.college==college)].mean()
                s=cao_data.dropna().points[(cao_data.college==college)].std()
                college_name.append(college); 
                mean.append(m); std.append(s)
 
    else:
        # If specified, rank for a specific year
        # else rank over all years
        if year>2000:
            name=str(year)+"no_aud"
            for college in cao_data['college']\
            [(cao_data['audition']==False)&(cao_data['year']==year)].dropna().unique():
                m=cao_data.dropna()['points']\
                [(cao_data['college']==college)&(cao_data['audition']==False)\
                 &(cao_data['year']==year)].mean()
                
                s=0
                college_name.append(college); 
                mean.append(m); std.append(s)

        else:
            name="all_years_no_aud"
            for college in cao_data['college']\
            [(cao_data['audition']==False)].dropna().unique():
                m=cao_data.dropna().points[(cao_data.college==college)\
                                                & (cao_data.audition==False)].mean()
                s=cao_data.dropna().points[(cao_data.college==college)\
                                                & (cao_data.audition==False)].std()
                college_name.append(college); 
                mean.append(m); std.append(s)

    rankings_df=pd.DataFrame({'college':college_name,'mean':mean,'std':std})
    rankings_df.columes=['college','mean','std']


    q="""SELECT college,mean,std FROM rankings_df ORDER BY mean DESC"""
    sorted_rankings=ps.sqldf(q,locals())
    sorted_rankings.columns=['college','mean','std']
    
    print(sorted_rankings)

    plot_college_rankings(sorted_rankings,name)
```

```python
# Function to get course rankings for a given year

def get_course_rankings(audition,year):
    course_name=[];mean=[];std=[];desc=[]

    if audition:

        if year>2000:
            for code in cao_data['code'][(cao_data['year']==year)].dropna().unique():
                m=cao_data['points'][(cao_data['code']==code)\
                                     & (cao_data['year']==year)].item()
                d=cao_data['desc'][(cao_data['code']==code)\
                                   & (cao_data['year']==year)].item()
                s=0
                course_name.append(code); mean.append(m); 
                std.append(s); desc.append(d)
        else:
            for code in cao_data['code'].dropna().unique():
                m=cao_data['points'][(cao_data['code']==code)].mean()
                s=cao_data['points'][(cao_data['code']==code)].std()
                d=cao_data['desc'][(cao_data['code']==code)].head(1).item()
                course_name.append(code); mean.append(m); 
                std.append(s); desc.append(d)

    else:

        if year>2000:
            for code in cao_data['code'][(cao_data['audition']==False)\
                                         & (cao_data['year']==year)].dropna().unique():
                m=cao_data['points'][(cao_data['code']==code)\
                                     & (cao_data['year']==year)].item()
                d=cao_data['desc'][(cao_data['code']==code)\
                                   & (cao_data['year']==year)].item()
                s=0
                course_name.append(code); mean.append(m); 
                std.append(s); desc.append(d)
        else:
            
            for code in cao_data['code'][(cao_data['audition']==False)].dropna().unique():
                m=cao_data['points'][(cao_data['code']==code)\
                                     & (cao_data['audition']==False)].mean()
                s=cao_data['points'][(cao_data['code']==code)\
                                     & (cao_data['audition']==False)].std()
                d=cao_data['desc'][(cao_data['code']==code)\
                                   & (cao_data['audition']==False)].head(1).item()
                course_name.append(code); mean.append(m); 
                std.append(s); desc.append(d)


    rankings_df=pd.DataFrame({'code':course_name,'desc':desc,'mean':mean,'std':std})
    rankings_df.columes=['code','desc','mean','std']
    
    q8="""SELECT * FROM rankings_df ORDER BY mean DESC LIMIT 25"""
    sorted_rankings=ps.sqldf(q8,locals())
    sorted_rankings.columns=['code','desc','mean','std']

    print(sorted_rankings.dropna())
    #plot_course_rankings(sorted_rankings)
```

```python
### PART 3: Time Series of Course Points

# Function to pull time series for individual courses
def course_time_series(df,course,printyn):
    
    prediction_year=df.year.max()+1

    time_series=df[['year','points']][(df['code']==course) &\
                                     (df['year']>=2009)]

    #Sort data by year descending
    time_series=time_series.sort_values(by='year', ascending=0)

    #Prediction through Linear Regression or ARIMA
    if len(time_series)>=5:

        if(printyn): print("Predictions "+course+" "+str(prediction_year))
        
        #Linear Regression on Time
        slope,intercept,r_value,p_valued,std_err=\
        do_regression(time_series.year,time_series.points,printyn,False)
        
        lin_reg_pred=prediction_year*slope+intercept
        

        #AR(1)
        slope,intercept,r_value,p_value,std_err_auto=\
        do_regression(time_series.points.values[1:],\
                      time_series.points.values[0:-1],printyn,False)
        
        auto_reg_pred=time_series.points.values[0]*slope+intercept
        

        #ARIMA(1,1,0)
        x1=time_series.points.values[0:-2]
        x2=time_series.points.values[1:-1]
        x3=time_series.points.values[2:]
        
        slope, intercept,r_value,p_value,std_err_ari=\
        do_regression((x2-x3),(x1-x2),printyn,False)
        
        ari_pred=(time_series.points.values[0]-\
                  time_series.points.values[1])*slope+\
        intercept+time_series.points.values[0]
  
        
        return lin_reg_pred,2*std_err,\
                auto_reg_pred,2*std_err_auto,\
                ari_pred,2*std_err_ari
            
    else:
        return 0,0,0,0,0,0
```

### PART 3: Predicting 2017 Points

```python
### PART 3.1: Testing models on 2016 points

# Training data 2001-2015
training_data=cao_data[cao_data['year']<2016]

# Test data 2016
test_data=cao_data[cao_data['year']==2016]
```

```python
f = csv.writer(open("predictions_2016.csv", "w+"))
f.writerow(["code","lin_reg","lin_reg_error^2",\
            "ar","ar_error^2",\
           "arima","arima_error^2"])

predictions=[]

# Loop over all courses in the training data
for course in test_data['code'].dropna():

    # If it was previously available run regression
    if any(training_data.code==course):
        lin_reg=0; ar=0; arima=0;
        lin_reg_err=0; ar_err=0; arima_err=0;
        
        lin_reg,lr_err,ar,ar_err,arima,arima_err=\
        course_time_series(training_data,course,False)
 
        # Print if sensible prediction values are returned
        if not any(np.isnan([lin_reg,ar,arima])) and \
        np.all(np.array([lin_reg,ar,arima])!=0):
           
            test_points=test_data['points'][test_data['code']==course].values[0]

            #Replace with Square-error
            lin_reg_err=(lin_reg-test_points)**2
            ar_err=(ar-test_points)**2
            arima_err=(arima-test_points)**2
            
            p=[course,round(lin_reg,0),round(lin_reg_err,0),\
              round(ar,0),round(ar_err,0),\
              round(arima,0),round(arima_err,0)]
            
            f.writerow(p)
            
            predictions.append({"code":course,\
                                "lin_reg":lin_reg,"lin_reg_err":lin_reg_err,\
                                "ar":ar,"ar_err":ar_err,\
                                "arima":arima,"arima_err":arima_err})
            


predictions=pd.DataFrame(predictions)
predictions=predictions.set_index('code')


# Compute MSE
print("Lin Reg error:\t",np.sqrt(predictions.lin_reg_err.dropna().mean()))
print("AR error:\t",np.sqrt(predictions.ar_err.dropna().mean()))
print("ARIMA error:\t",np.sqrt(predictions.arima_err.dropna().mean()))


```
